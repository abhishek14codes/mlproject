```python
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['average']=df['total_score']/3
df.head()
```

```python
reading_full=df[df['reading_score']==100].shape[0]
writing_full=df[df['writing_score']==100].shape[0]
math_full=df[df['math_score']==100].shape[0]
print(f"students with full reading score : {reading_full}")
print(f"students with full writing score : {writing_full}")
print(f"students with full math score : {math_full}")
```

```python
reading_lessthan_20=df[df['reading_score']<=20].shape[0]
writing_lessthan_20=df[df['writing_score']<=20].shape[0]
math_lessthan_20=df[df['math_score']<=20].shape[0]
print(f"students with less than 20 reading score : {reading_lessthan_20}")
print(f"students with less than 20 writing score : {writing_lessthan_20}")
print(f"students with less than 20 math score : {math_lessthan_20}")
```
